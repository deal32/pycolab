```
import torch
```

```
import torchvision
```

```
mnist_train = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST(
        root='/tmp',
        train=True,
        download=True,
        transform=torchvision.transforms.ToTensor()
), 
    batch_size=10
)
```

```
import torch.nn as nn

SHAG_OBUCHENIA = 0.001
EPOCH = 5




#y = w * x + b
sloy_1 = nn.Linear(in_features=(784),out_features=(666))
sloy_2 = nn.Linear(in_features=(666),out_features=(10))
activation = nn.Sigmoid()
softmax = nn.Softmax()

loss = nn.CrossEntropyLoss()
optimizator = torch.optim.Adam([sloy_1.weight,sloy_1.bias,sloy_2.weight,sloy_2.bias] , lr = SHAG_OBUCHENIA)
```

```
import matplotlib.pyplot as plt

losses = []


for i in range(EPOCH):
    for iteration, data in enumerate(mnist_train):
        #print(data[0].shape,data[1])
        #plt.imshow(data[0][0].permute(1,2,0).squeeze().numpy())

        optimizator.zero_grad()

        x = data[0] ## Данные на вход (изображения)
        y = data[1] ## Подписи изображений (метки)

        y_pred = activation(sloy_1(torch.reshape(x,(-1,784))))
        y_pred = activation(sloy_2(y_pred))
        y_prop = softmax(y_pred)
        
        nomer_classa = torch.argmax(y_prop,dim=1)
        
        tochnost = torch.sum(y == nomer_classa, dtype=torch.float) / BATCH_SIZE

        loss_val = loss(y_prop,y)
        
        if iteration % 100 == 0:
            print('Эпоха {} Итерация - {}, значение потери - {}, точность - {}'.format(i,iteration,loss_val,tochnost))

        #Вычисление градиентов
        loss_val.backward()

        #Обновление весов (т.е. применение весов)
        optimizator.step()

        losses.append(loss_val.item())
    print('Конец эпохи')
```

```
import torch  # Подключение библиотеки PyTorch  для ненйронных сетей
import torchvision # Подключение библиотеки для загрузки датасетов
import torch.nn as nn # Подключение  модуля библиотеки для слоев
import matplotlib 
import matplotlib.pyplot as plt # Подключение библиотеки для построения графиков
%matplotlib inline  # Необходимо писать, чтобы графики строились в ячейке

BATCH_SIZE = 10
SHAG_OBUCHENIA = 0.001
EPOCH = 4
```

```
#Датасет для обучения
mnist_train = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST(
        root='/tmp',
        train=True,
        download=True,
        transform=torchvision.transforms.ToTensor()
), 
    batch_size=BATCH_SIZE
)

#Датасет для тестов
mnist_test = torch.utils.data.DataLoader(
    torchvision.datasets.MNIST(
        root='/tmp',
        train=False,
        download=True,
        transform=torchvision.transforms.ToTensor()
), 
    batch_size=BATCH_SIZE
)
```

### Наша модель

```
#y = w * x + b
sloy_1 = nn.Linear(in_features=(784),out_features=(666))
sloy_2 = nn.Linear(in_features=(666),out_features=(10))
activation = nn.Sigmoid()
softmax = nn.Softmax()

loss = nn.CrossEntropyLoss()
optimizator = torch.optim.Adam([sloy_1.weight,sloy_1.bias,sloy_2.weight,sloy_2.bias] , lr = SHAG_OBUCHENIA)
```

```
losses_train = [] # Массив со средним значением потери в эпохе на этапе обучения
losses_test = [] # Массив со средним значением потери в эпохе на этапе тестирования
massiv_tochnosti_train = [] # Массив со средним значением точности в эпохе на этапе обучения
massiv_tochnosti_test = [] # Массив со средним значением точности в эпохе на этапе тестирования

for i in range(EPOCH):
    losses_epoch = []   
    tochnosti_epoch = []  
    #Цикл для обучения
    for iteration, data in enumerate(mnist_train):

        
        optimizator.zero_grad()

        x = data[0] ## Данные на вход (изображения)
        y = data[1] ## Подписи изображений (метки)

        y_pred = activation(sloy_1(torch.reshape(x,(-1,784))))
        y_pred = activation(sloy_2(y_pred))
        y_prop = softmax(y_pred)
        
        nomer_classa = torch.argmax(y_prop,dim=1)
        
        tochnost = torch.sum(y == nomer_classa, dtype=torch.float) / BATCH_SIZE

        loss_val = loss(y_prop,y)
        
        if iteration % 100 == 0:
            print('Эпоха {} Итерация - {}, значение потери - {}, точность - {}'.format(i,iteration,loss_val,tochnost))

        #Вычисление градиентов
        loss_val.backward()

        #Обновление весов (т.е. применение весов)
        optimizator.step()

        losses_epoch.append(loss_val.item())
        tochnosti_epoch.append(tochnost)
    #Добавление в массив среднего значения точности за текущую эпоху
    massiv_tochnosti_train.append(torch.sum(torch.as_tensor(tochnosti_epoch)) / len(tochnosti_epoch))
    #Добавление в массив среднего значения потери за текущую эпоху
    losses_train.append(torch.sum(torch.as_tensor(losses_epoch)) / len(losses_epoch))   
    
    
    print('Конец эпохи')
```

### Transfer Learning (Обучение с переносом)

```
import torchvision
import torch
import torchvision.models as models
vgg = models.vgg16_bn(pretrained=True)
print(vgg)
```

В коде ниже выполняется загрузка картинки по **ссылке**. Вы можете ее заменять по своему усмотрению

```
from PIL import Image
import requests
from io import BytesIO
import json

import matplotlib.pyplot as plt
%matplotlib inline

classes_str = requests.get("https://s3.amazonaws.com/deep-learning-models/image-models/imagenet_class_index.json")
classes_obj = classes_str.json()

URL_CAR = "https://d17fnq9dkz9hgj.cloudfront.net/breed-uploads/2018/08/american-staffordshire-terrier-detail.jpg?bust=1535565083&width=355"
response = requests.get(URL_CAR)
img = Image.open(BytesIO(response.content))
plt.imshow(img)
```

Предобученные модели  требуют те же способы обработки что и при обучении. Они перечислены [тут](https://pytorch.org/docs/stable/torchvision/models.html)

```
resize = torchvision.transforms.Resize((224,224))
totensor = torchvision.transforms.ToTensor()
normalize = torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

#img_explain = [
#    [0], #Красный [0-255]  -> [0. - 1.] -> (x_r - mean[0]) / std[0]
#    [0], #Зеленый [0-255]  -> [0. - 1.] -> (x_g - mean[1]) / std[1]
#    [0] #Синий [0-255]  -> [0. - 1.] -> (x_b - mean[2]) / std[2]
#]

#Предобработка
transformed_img = normalize(totensor(resize(img)))

#Вывод - что на картинке?
# По размерности на вход ндо дать 4хмерный тензор [1,224,224,3]
ans = vgg.forward(transformed_img[None,:])
ans = torch.nn.functional.softmax(ans)
ans.shape
```

```
ans
```

```
#topk - возращает k (k = 5) пар - (значение_вероятности, номер индекса)
a1 = torch.topk(ans[0],5)
#zip() - объединяет два массива и возвращаетих пару, которую можно использовать в цикле
for prob, ind in zip(a1[0],a1[1]):
    print(prob.item(),classes_obj[str(ind.item())])
```

### Редактирование сети

Нам необходимо заменить количество нейронов на выходе в последнем слое

```
vgg.classifier[6] = torch.nn.Linear(in_features=4096,out_features=2)
vgg
```

```
train_loader = torch.utils.data.DataLoader(train_dataset,batch_size=10,shuffle=True)
```

```
#Переводит модель в режим обучения
vgg.train()

device = torch.device('cuda:0')
vgg.to(device)

criteria = torch.nn.CrossEntropyLoss()
optimizator = torch.optim.Adam(vgg.parameters())
softmax = torch.nn.Softmax()

for idx,data in enumerate(train_loader):
    optimizator.zero_grad()
    
    data_gpu = data[0].to(device)
    label_gpu = data[1].to(device)
    hotdog_output = vgg.forward(data_gpu) 
    y = softmax(hotdog_output)
    loss = criteria(y,label_gpu)
    
    print(idx,loss.item())
    loss.backward()
    optimizator.step()
```
